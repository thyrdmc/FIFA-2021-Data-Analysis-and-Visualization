### **Data Cleaning**

```python
# Filling the Missing Variables Values for Proper Data Visualization

df['age'].fillna(df['age'].mean(), inplace = True)
df['dob'].fillna(df['dob'][0], inplace = True)
df['height_cm'].fillna(df['height_cm'].mean(), inplace = True)
df['weight_kg'].fillna(df['weight_kg'].mean(), inplace = True)
df['nationality'].fillna(df['nationality'][0], inplace = True)
df['club_name'].fillna(df['club_name'][3000], inplace = True)
df['league_name'].fillna(df['league_name'][3000], inplace = True)
df['league_rank'].fillna(df['league_rank'][3000], inplace = True)
df['overall'].fillna(df['overall'].min(), inplace = True)
# df['player_positions'].fillna(df['player_positions'][3000], inplace = True)
df['preferred_foot'].fillna(df['preferred_foot'][3000], inplace = True)
df['international_reputation'].fillna(df['international_reputation'].mean(), inplace = True)
df['weak_foot'].fillna(df['weak_foot'].mean(), inplace = True)
df['skill_moves'].fillna(df['skill_moves'].mean(), inplace = True)
df['work_rate'].fillna(df['work_rate'][3000], inplace = True)
df['body_type'].fillna(df['body_type'][3000], inplace = True)
df['real_face'].fillna(df['real_face'][3000], inplace = True)
df['release_clause_eur'].fillna(df['release_clause_eur'].mean(), inplace = True)
df['player_tags'].fillna(df['player_tags'][7000], inplace = True)
df['team_position'].fillna(df['team_position'][3], inplace = True)
df['team_jersey_number'].fillna(df['team_jersey_number'][3], inplace = True)
df['loaned_from'].fillna(df['loaned_from'][3], inplace = True)
df['joined'].fillna(df['joined'][3], inplace = True)
df['contract_valid_until'].fillna(df['contract_valid_until'][3000], inplace = True)
df['nation_position'].fillna(df['nation_position'][5], inplace = True)
df['nation_jersey_number'].fillna(df['nation_jersey_number'][5], inplace = True)
df['pace'].fillna(df['pace'].mean(), inplace = True)
df['shooting'].fillna(df['shooting'].mean(), inplace = True)
df['passing'].fillna(df['passing'].mean(), inplace = True)
df['dribbling'].fillna(df['dribbling'].mean(), inplace = True)
df['defending'].fillna(df['defending'].mean(), inplace = True)
df['physic'].fillna(df['physic'].mean(), inplace = True)
df['player_traits'].fillna(df['player_traits'][3000], inplace = True)
```

```python
from math import pi

idx = 1
plt.figure(figsize=(15,45))
for position_name, features in df.groupby(df['team_position'])[player_features].mean().iterrows():
    top_features = dict(features.nlargest(5))
    
    # Number of Variable
    categories=top_features.keys()
    N = len(categories)

    # We will plot the first row of the data frame.
    values = list(top_features.values())
    values += values[:1]

    # Calculating Angle of Each Axis in the Plot (Plot / Number of Variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    ax = plt.subplot(9, 3, idx, polar=True)

    plt.xticks(angles[:-1], categories, color='grey', size=8)

    ax.set_rlabel_position(0)
    plt.yticks([25,50,75], ["25","50","75"], color="grey", size=7)
    plt.ylim(0,100)
    
    plt.subplots_adjust(hspace = 0.5)
    
    # Plot Data
    ax.plot(angles, values, linewidth=1, linestyle='solid')

    # Fill Area
    ax.fill(angles, values, 'b', alpha=0.1)
    
    plt.title(position_name, size=11, y=1.1)

    idx += 1

    if(idx == 28):
      break;
plt.savefig("Plots/Relationship_Between_Player_Positions&Features.png")

     
```
